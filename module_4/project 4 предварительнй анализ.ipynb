```python
#4.4
select 
    min(scheduled_departure)
from 
    dst_project.flights
    
    
select 
    EXTRACT('day' from max(scheduled_arrival-scheduled_departure))*1440+
    EXTRACT('hour' from max(scheduled_arrival-scheduled_departure))*60+
    EXTRACT('minute' from max(scheduled_arrival-scheduled_departure))
from 
    dst_project.flights

    
select 
    EXTRACT('day' from avg(scheduled_arrival-scheduled_departure))*1440+
    EXTRACT('hour' from avg(scheduled_arrival-scheduled_departure))*60+
    EXTRACT('minute' from avg(scheduled_arrival-scheduled_departure))
from 
    dst_project.flights


    
with a as
(select 
    flight_id,
    scheduled_arrival-scheduled_departure as delta_time,
    departure_airport,
    arrival_airport
from dst_project.flights
)
select 
    f.departure_airport,
    f.arrival_airport
from 
    dst_project.flights f
    join a on f.flight_id=a.flight_id
order by 
    delta_time desc
limit 1

```
